```python
#since the histogram predicted that some graphs dont look like gausian bell curve. our data is skewed
df['total_rooms']=np.log(df['total_rooms']+1) #+1 so that values dont go to 0
df['total_bedrooms']=np.log(df['total_bedrooms']+1)
df['population']=np.log(df['population']+1)
df['households']=np.log(df['households']+1)
```

```python
h=pd.get_dummies(df["ocean_proximity"])#string to 1 or 0 . same as onehotencoding 
h.head()
```

```python
df=df.join(pd.get_dummies(df["ocean_proximity"])).drop(["ocean_proximity"],axis=1)
```

```python
df.columns
```

```python
X=df[['longitude', 'latitude', 'housing_median_age', 'total_rooms',
  'total_bedrooms', 'population', 'households', 'median_income',
       '<1H OCEAN', 'INLAND', 'ISLAND', 'NEAR BAY',
       'NEAR OCEAN']]
y=df['median_house_value']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.40, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train)
```

```python
y_pred=regression.predict(X_test)
```

```python
df=pd.DataFrame({'actual':y_test,'prediction':y_pred})
df.head()
```

```python
regression.score(X_test,y_test)
```

```python
sns.distplot((y_test-y_pred),bins=50)
```

```python
plt.scatter(y_test,y_pred)
plt.ylabel("actual house value")
plt.xlabel("predicted house value")
```
